```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
d=pd.read_csv('insurance.csv')
d
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
d.loc[:,['smoker']]=le.fit_transform(d.loc[:,['smoker']])
d.loc[:,['sex']]=le.fit_transform(d.loc[:,['sex']])
d.loc[:,['region']]=le.fit_transform(d.loc[:,['region']])
```

```python
d
```

```python
X=d[['age', 'sex', 'bmi', 'children', 'smoker','region']]
```

```python
y=d[['charges']]
```

```python
from scipy.stats import shapiro
shapiro(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from pycaret.regression import *
```

```python
reg=setup(data=d,target='charges')
```

```python
compare_models()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gb = GradientBoostingRegressor(alpha=0.9, ccp_alpha=0.0, criterion='friedman_mse',
                          init=None, learning_rate=0.1, loss='ls', max_depth=3,
                          max_features=None, max_leaf_nodes=None,
                          min_impurity_decrease=0.0, min_impurity_split=None,
                          min_samples_leaf=1, min_samples_split=2,
                          min_weight_fraction_leaf=0.0, n_estimators=100,
                          n_iter_no_change=None, presort='deprecated',
                          random_state=6846, subsample=1.0, tol=0.0001,
                          validation_fraction=0.1, verbose=0, warm_start=False)
```

```python
gb.fit(X_train,y_train)
```

```python
predict=gb.predict(X_train)
```

```python
predict1=gb.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmsetrain=np.sqrt(mean_squared_error(predict,y_train))
rmsetrain
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(predict,y_train)
```

```python
r2_score(predict1,y_test)
```
